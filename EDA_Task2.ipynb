```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
'''Questions Before Analysis

1. How many rows and columns are present?
2. What types of variables exist (numeric, categorical)?
3. Are there missing values?
4. Are there outliers in numeric features?
5. How are variables distributed?
6. Are there relationships between variables?
7. Are there any unusual patterns or anomalies? '''
```

```python
# 2 - Check Missing Values
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# 3 - Analyze Distributions
df.hist(figsize=(15,10))
plt.show()
```

```python
sns.histplot(df.select_dtypes(include=np.number).iloc[:,0], kde=True)
plt.show()
```

```python
# 4 - Detect Outliers
plt.figure(figsize=(10,5))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.show()
```

```python
# 5 - Detect Data Issues
"""Potential Data Issues

- Missing values
- Outliers
- Skewed distributions
- Incorrect data types
- Duplicate records
- Imbalanced categories
"""

df.duplicated().sum()
```
